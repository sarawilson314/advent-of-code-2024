```python
def find_guard_line_index(data: list[str]) -> int:
    for line in range(len(data)):
        if '^' in data[line]:
            guard_line_index = line
    return guard_line_index

def find_guard_element_index(data: list[str], guard_line_index: str) -> int:
    for element in range(len(data[guard_line_index])):
        if data[guard_line_index][element] == '^':
            guard_element_index = element
    return guard_element_index

def mark_guard_position_and_move_guard_to_next_position(data: list[str],
                                                        guard_has_exited: bool,
                                                        guard_direction: str,
                                                        guard_line_index: int,
                                                        guard_element_index: int) -> int:
    # Mark the guard's current position with an X
    data[guard_line_index] = (
    data[guard_line_index][:guard_element_index] + 'X' + data[guard_line_index][guard_element_index+1:]
    )
    print(f'Applied X to data line {guard_line_index} element {guard_element_index}')
    # Move the guard based on his direction
    if guard_direction == '^':
        if guard_line_index - 1 < 0:
            guard_has_exited = True
        if guard_has_exited == False:
            if data[guard_line_index - 1][guard_element_index] != '#':
                guard_line_index -= 1
            else:
                guard_direction = '>'
    elif guard_direction == '>':
        if guard_element_index + 1 >= len(data[guard_line_index]):
            guard_has_exited = True
        if guard_has_exited == False:
            if data[guard_line_index][guard_element_index + 1] != '#':
                guard_element_index += 1
            else:
                guard_direction = 'v'
    elif guard_direction == 'v':
        if guard_line_index + 1 >= len(data):
            guard_has_exited = True
        if guard_has_exited == False:
            if data[guard_line_index + 1][guard_element_index] != '#':
                guard_line_index += 1
            else:
                guard_direction = '<'
    elif guard_direction == '<':
        if guard_element_index - 1 < 0:
            guard_has_exited = True
        if guard_has_exited == False:
            if data[guard_line_index][guard_element_index - 1] != '#':
                guard_element_index -= 1
            else: guard_direction = '^'
    return data, guard_has_exited, guard_direction, guard_line_index, guard_element_index

guard_line_index = find_guard_line_index(data=data)
guard_element_index = find_guard_element_index(data=data, guard_line_index=guard_line_index)
guard_direction = data[guard_line_index][guard_element_index]
print(guard_line_index)
print(guard_element_index)
print(guard_direction)

def follow_guard(data: list[str], guard_direction: str, guard_line_index: int, guard_element_index: int):
    guard_has_exited = False
    while (guard_line_index in range(len(data))) and (guard_element_index in range(len(data[guard_line_index]))):
        while guard_has_exited == False:
            print(f'Guard Line: {guard_line_index}')
            print(f'Guard Element: {guard_element_index}')
            print(f'Guard Direction: {guard_direction}')
            data, guard_has_exited, guard_direction, guard_line_index, guard_element_index = mark_guard_position_and_move_guard_to_next_position(
                data=data,
                guard_has_exited=guard_has_exited,
                guard_direction=guard_direction,
                guard_line_index=guard_line_index,
                guard_element_index=guard_element_index)
        print("Guard has exited")
        break
    return data

for line in data:
    print(line)

new_data = follow_guard(
    data=data.copy(),
    guard_direction=guard_direction,
    guard_line_index=guard_line_index,
    guard_element_index=guard_element_index)

num_of_x = 0
for line in new_data:
    print(line)
    for element in line:
        if element == 'X':
            num_of_x +=1

print(f"Result: {num_of_x} X's")
```

It doesn't really matter what you choose to use as an obstacle so long as you and The Historians can put it into position without the guard noticing. The important thing is having enough options that you can find one that minimizes time paradoxes, and in this example, there are 6 different positions you could choose.

You need to get the guard stuck in a loop by adding a single new obstruction. How many different positions could you choose for this obstruction?


```python
def mark_guard_position_and_move_guard_to_next_position(data: list[str],
                                                        guard_has_exited: bool,
                                                        guard_direction: str,
                                                        guard_symbol: str,
                                                        infinite_loop: bool,
                                                        past_guard_positions: set,
                                                        guard_line_index: int,
                                                        guard_element_index: int) -> int:
    # Mark the guard's current position
    data[guard_line_index] = (
    data[guard_line_index][:guard_element_index] + guard_symbol + data[guard_line_index][guard_element_index+1:]
    )
    current_position = (guard_line_index, guard_element_index, guard_direction)
    if current_position in past_guard_positions:
        infinite_loop = True
    else:
        past_guard_positions.add(current_position)
    #print(f'Current Guard Position: {current_position}')
    #print(f'Past Guard Positions: {past_guard_positions}')
    #print(f'Guard Line Index: {guard_line_index}')
    #print(f'Guard Element Index: {guard_element_index}')
    #print(f'/n')
    # Move the guard based on his direction
    if guard_direction == '^':
        if guard_line_index - 1 < 0:
            guard_has_exited = True
        if guard_has_exited == False:
            if data[guard_line_index - 1][guard_element_index] != '#':
                if data[guard_line_index - 1][guard_element_index] == '-':
                    guard_symbol = '+'
                else:
                    guard_symbol = '|'
                guard_line_index -= 1
            else:
                guard_direction = '>'
                guard_symbol = '+'
    elif guard_direction == '>':
        if guard_element_index + 1 >= len(data[guard_line_index]): # must be >= since indexing starts at 0
            guard_has_exited = True
        if guard_has_exited == False:
            if data[guard_line_index][guard_element_index + 1] != '#':
                if data[guard_line_index][guard_element_index + 1] == '|':
                    guard_symbol = '+'
                else:
                    guard_symbol = '-'
                guard_element_index += 1
            else:
                guard_direction = 'v'
                guard_symbol = '+'
    elif guard_direction == 'v':
        if guard_line_index + 1 >= len(data): # must be >= since indexing starts at 0
            guard_has_exited = True
        if guard_has_exited == False:
            if data[guard_line_index + 1][guard_element_index] != '#':
                if data[guard_line_index + 1][guard_element_index] == '-':
                    guard_symbol = '+'
                else:
                    guard_symbol = '|'
                guard_line_index += 1
            else:
                guard_direction = '<'
                guard_symbol = '+'
    elif guard_direction == '<':
        if guard_element_index - 1 < 0:
            guard_has_exited = True
        if guard_has_exited == False:
            if data[guard_line_index][guard_element_index - 1] != '#':
                if data[guard_line_index][guard_element_index - 1] == '|':
                    guard_symbol = '+'
                else:
                    guard_symbol = '-'
                guard_element_index -= 1
            else:
                guard_direction = '^'
                guard_symbol = '+'
    return data, guard_has_exited, guard_direction, guard_symbol, infinite_loop, past_guard_positions, guard_line_index, guard_element_index

def follow_guard(data: list[str], guard_direction: str, guard_line_index: int, guard_element_index: int, num_times_guard_stuck_in_infinite_loop: int):
    guard_has_exited = False
    guard_symbol = '|'
    infinite_loop = False
    past_guard_positions = set()
    while (guard_line_index in range(len(data))) and (guard_element_index in range(len(data[guard_line_index]))):
        while not guard_has_exited and not infinite_loop:
            #for line in data:
            #    print(line)
            data, guard_has_exited, guard_direction, guard_symbol, infinite_loop, past_guard_positions, guard_line_index, guard_element_index = mark_guard_position_and_move_guard_to_next_position(
                data=data,
                guard_has_exited=guard_has_exited,
                guard_direction=guard_direction,
                guard_symbol=guard_symbol,
                infinite_loop=infinite_loop,
                past_guard_positions=past_guard_positions,
                guard_line_index=guard_line_index,
                guard_element_index=guard_element_index)
        if guard_has_exited:
            print("Guard has exited")
        if infinite_loop:
            print("Guard is stuck in an infinite loop")
            num_times_guard_stuck_in_infinite_loop += 1
        break
    return num_times_guard_stuck_in_infinite_loop, data

guard_line_index = find_guard_line_index(data=data)
guard_element_index = find_guard_element_index(data=data, guard_line_index=guard_line_index)
guard_direction = data[guard_line_index][guard_element_index]
num_times_guard_stuck_in_infinite_loop = 0

#for line_index in range(len(data)):
#    for element_index in range(len(data[line_index])):
#        new_data = data.copy()
#        new_data[line_index] = (new_data[line_index][:element_index] + "#" + new_data[line_index][element_index+1:])
#        #print(new_data)

for line_index in range(len(data)):
#for line_index in range(5, 7):
    for element_index in range(len(data[line_index])):
        new_data = data.copy()
        new_data[line_index] = new_data[line_index][:element_index] + "#" + new_data[line_index][element_index+1:]
        #print(f'Test # applied at Line: {line_index}, Element: {element_index}')
        num_times_guard_stuck_in_infinite_loop, new_data = follow_guard(
            data=new_data,
            guard_direction=guard_direction,
            guard_line_index=guard_line_index,
            guard_element_index=guard_element_index,
            num_times_guard_stuck_in_infinite_loop=num_times_guard_stuck_in_infinite_loop)
        #print(f"Line: {line_index}, Element: {element_index}")
        
print(f"""Number of times guard stuck in infinite loop: {num_times_guard_stuck_in_infinite_loop}""")
```
