The unusual data (your puzzle input) consists of many reports, one report per line. Each report is a list of numbers called levels that are separated by spaces. For example:

7 6 4 2 1
1 2 7 8 9
9 7 6 2 1
1 3 2 4 5
8 6 4 4 1
1 3 6 7 9

This example data contains six reports each containing five levels.

The engineers are trying to figure out which reports are safe. The Red-Nosed reactor safety systems can only tolerate levels that are either gradually increasing or gradually decreasing. So, a report only counts as safe if both of the following are true:

    The levels are either all increasing or all decreasing.
    Any two adjacent levels differ by at least one and at most three.

In the example above, the reports can be found safe or unsafe by checking those rules:

    7 6 4 2 1: Safe because the levels are all decreasing by 1 or 2.
    1 2 7 8 9: Unsafe because 2 7 is an increase of 5.
    9 7 6 2 1: Unsafe because 6 2 is a decrease of 4.
    1 3 2 4 5: Unsafe because 1 3 is increasing but 3 2 is decreasing.
    8 6 4 4 1: Unsafe because 4 4 is neither an increase or a decrease.
    1 3 6 7 9: Safe because the levels are all increasing by 1, 2, or 3.

So, in this example, 2 reports are safe.

Analyze the unusual data from the engineers. How many reports are safe?


```python
import pandas as pd

record1 = [7, 6, 4, 2, 1]
record2 = [1, 2, 7, 8, 9]
record3 = [9, 7, 6, 2, 1]
record4 = [1, 3, 2, 4, 5]
record5 = [8, 6, 4, 4, 1]
record6 = [1, 3, 6, 7, 9]
```

```python
def safety_check(record):
    record_length = len(record)
    initial_index = record_length - 1
    
    index = initial_index
    next_index = index - 1
    safety_asc = 'Safe'
    while next_index >= 0 and safety_asc == 'Safe':
        if record[index] > record[next_index]:
            index -= 1
            next_index -= 1
        else:
            safety_asc = 'Not Safe'
    
    index = initial_index
    next_index = index - 1
    safety_desc = 'Safe'
    while next_index >= 0 and safety_desc == 'Safe':
        if record[index] < record[next_index]:
            index -= 1
            next_index -= 1
        else:
            safety_desc = 'Not Safe'

    if safety_asc == 'Safe' or safety_desc == 'Safe':
        safety = 'Safe'
    else:
        safety = 'Not Safe'

    if safety == 'Safe':
        index = initial_index
        next_index = index - 1
        while next_index >= 0 and safety == 'Safe':
            if (abs(record[index] - record[next_index]) >= 1) & (abs(record[index] - record[next_index]) <= 3):
                index -= 1
                next_index -= 1
            else:
                safety = 'Not Safe'

    return(safety)

result1 = safety_check(record1)
result2 = safety_check(record2)
result3 = safety_check(record3)
result4 = safety_check(record4)
result5 = safety_check(record5)
result6 = safety_check(record6)

print(f'Record 1 is {result1}')
print(f'Record 2 is {result2}')
print(f'Record 3 is {result3}')
print(f'Record 4 is {result4}')
print(f'Record 5 is {result5}')
print(f'Record 6 is {result6}')
```

```python
data = pd.read_table('day-2-input.txt', delimiter='  ', header=None, engine='python')

count_of_safes = 0
for row in data[0]:
    row = row.split(' ')
    row = [int(x) for x in row]
    safety = safety_check(row)
    if safety == 'Safe':
        count_of_safes += 1
    print(f'Row {row} is {safety}')

print("Count of Safe Records:", count_of_safes)
```
